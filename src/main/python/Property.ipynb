```python
#%matplotlib notebook

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.colors import Normalize
from matplotlib import cm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter
plt.rcParams['date.epoch'] = '0000-12-31'
import geopandas

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
data = pd.read_csv(r'[DATA_PATH]/price_evolutions.csv')
```

```python
Writer = animation.writers['ffmpeg']
writer = animation.FFMpegWriter(fps=10)

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(20, 10, forward=True)

ax2.set_xlim(0, len(data.y.unique()))
ax2.set_ylim(0, len(data.x.unique()))

dates = data.date.unique()

def animate(i):
    ax1.cla() 
    ax2.cla() 
    x = data[data.date <= dates[i]].groupby(['date', 'x', 'y']).g.mean().unstack('date').fillna(1).T.cumprod()
    
    x.apply(lambda row: np.median(row), axis=1).plot(ax = ax1, color='red')
    x.apply(lambda row: np.mean(row), axis=1).plot(ax = ax1, color='blue')

    x.apply(lambda row: np.quantile(row, 0.01), axis=1).plot(ax = ax1, color='red', alpha=0.05)
    x.apply(lambda row: np.quantile(row, 0.99), axis=1).plot(ax = ax1, color='red', alpha=0.05)

    x.apply(lambda row: np.quantile(row, 0.05), axis=1).plot(ax = ax1, color='red', alpha=0.1)
    x.apply(lambda row: np.quantile(row, 0.95), axis=1).plot(ax = ax1, color='red', alpha=0.1)

    x.apply(lambda row: np.quantile(row, 0.1), axis=1).plot(ax = ax1, color='red', alpha=0.25)
    x.apply(lambda row: np.quantile(row, 0.9), axis=1).plot(ax = ax1, color='red', alpha=0.25)

    x.apply(lambda row: np.quantile(row, 0.25), axis=1).plot(ax = ax1, color='red', alpha=0.5)
    x.apply(lambda row: np.quantile(row, 0.75), axis=1).plot(ax = ax1, color='red', alpha=0.5)

    h = data[data.date == dates[i]].groupby(['x', 'y']).g.mean().unstack('y').values
    ax2.contourf(h, cmap=cm.RdYlGn, extend='both', vmin=0.975, vmax=1.025)
    ax2.axis('off')
    ax2.set_title(dates[i])
    ax1.set_title('Price Evolution')
    ax1.spines['right'].set_visible(False)
    ax1.spines['top'].set_visible(False)
    
ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(dates), repeat=True)
ani.save('property_evolution.mp4', writer=writer, dpi=200)
```
